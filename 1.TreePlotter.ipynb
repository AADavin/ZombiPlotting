```python
from IPython.display import SVG, display

import svgwrite
import random
import ete3
import random
import os
import seaborn as sns
import drawSvg as draw
import pandas as pd


def show_svg(url):
    display(SVG(url))
    
from ZombiPlotter import SpeciesTreeDrawer 
```

```python
# We read the constraints
df = pd.read_csv("Bootstrap_528.tsv", sep="\t", names = "dn,rc,bs,wt".split(","))
sdf = df[df["bs"] >= 95]
```

```python
with open("MaxTiC528Tree.nwk") as f:
    t = ete3.Tree(f.readline(), format=1)    
c1, c2 = t.get_children()
group1 = {n.name for n in c1.traverse()}
group2 = {n.name for n in c2.traverse()}
```

```python
treefile = "Test528/T/CompleteTree.nwk"
eventsfile = "Test528/T/Events.tsv"

std = SpeciesTreeDrawer(treefile, eventsfile, 600, 1350, background="black")

d = std.DrawInner("./Tree528.svg", stroke="red", stroke_width=2)

for index, row in sdf.sort_values("wt", ascending=False)[1500:1800][["dn", "rc"]].iterrows():
    dn = str(row["dn"])
    rc = str(row["rc"])
    #print(dn,rc)
    #wt = row["wt"] / 10
    #print(dn, rc)
    if dn in group1 and rc in group2:
        d = std.draw_constraint("./Tree528.svg",dn,rc, "white" , 2)
    elif dn in group2 and rc in group1:
         d = std.draw_constraint("./Tree528.svg",dn,rc, "cyan", 2)
    else:        
        d = std.draw_constraint("./Tree528.svg",dn,rc, "green", 2)

std.d
#std.d.savePng('100MostSupportedConstraints.png')
```
