```python
from IPython.display import SVG, display

import svgwrite
import random
import ete3
import random
import os
import seaborn as sns
import drawSvg as draw
import pandas as pd


def show_svg(url):
    display(SVG(url))
    
from ZombiPlotter import SpeciesTreeDrawer 
```

```python
# We read the constraints
df = pd.read_csv("HighSupportTransfer_528.tsv", sep="\t", names = "dn,rc,wt".split(","))
```

```python
ds = pd.read_csv("TableSpecies.tsv", sep="\t")
```

```python
sc2phylum = dict(zip(ds["ShortCode"], ds["GTDB Taxonomy"].apply(lambda x: x.split(";")[1])))
```

```python
with open("MaxTiC528Tree.nwk") as f:
    t = ete3.Tree(f.readline(), format=1)    
c1, c2 = t.get_children()
group1 = {n.name for n in c1.traverse()}
group2 = {n.name for n in c2.traverse()}
```

```python
# We prepare the mapping of names:

phylum2number = dict()

for n in t.traverse("postorder"):
    
    if n.is_leaf():
        phylum = sc2phylum[n.name]
        n.add_feature("phylum", {phylum})
        if phylum not in phylum2number:
            phylum2number[phylum] = 0
        phylum2number[phylum] += 1
    else:
        c1, c2 = n.get_children()        
        n.add_feature("phylum", c1.phylum | c2.phylum)
        
map_names = dict()

for n in t.traverse():
    if len(n.phylum) == 1 and phylum2number[list(n.phylum)[0]] == len(n):
        if len(n) == 1:
            continue
        map_names[n.name] = list(n.phylum)[0][3:]
print("Done!")
```

```python
# This draws the basic tree
treefile = "Test528/T/CompleteTree.nwk"
eventsfile = "Test528/T/Events.tsv"

std = SpeciesTreeDrawer(treefile, eventsfile, 600, 1350, background="white")
d = std.DrawInner("./Tree528.svg", stroke="black", stroke_width=2)
std.add_names_advanced("./Tree528.svg", map_names = map_names, colour="green", y_adj=+10, x_adj=-50, font_size = 15)
std.d
```

```python
def rgb2hex(r,g,b):
    return "#{:02x}{:02x}{:02x}".format(r,g,b)
```
